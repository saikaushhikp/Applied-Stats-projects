```python
#plot for bar graphs 
def BarGraph(title, xlabel, ylabel, x_axis, y_axis) :
    
    plt.figure(figsize=(6, 5))
    no_of_country =[y_axis.count(category) for category in x_axis]
    plt.bar(x_axis,no_of_country,color = 'deepskyblue', width = 0.5, zorder = 2)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.grid(axis = 'y', linestyle = '-.', alpha = 0.5, zorder = 1)
    #plt.savefig(title+".svg")
    plt.show()
    

#puchasing power aspect
BarGraph("Purchasing Power Category of Countries", "Purchasing Power Category", "Number of Countries", levels, purchasing_power_category)

#Safety aspect
BarGraph("Safety Category of Countries", "Safety Category", "Number of Countries", levels, safety_category)

#health care aspect
BarGraph("Health Care Category of Countries", "Health Care Category", "Number of Countries", levels, health_care_category)

#Cost of Living aspect
BarGraph("Cost of Living Category of Countries", "Cost of Living Category", "Number of Countries", levels, col_category)

#pollution aspect
BarGraph("Pollution Category of Countries", "Pollution Category", "Number of Countries", levels, pollution_category)
```

```python
# polt for Ogives
def OgiveGraph(title, xlabel, ylabel, x_axis, y_axis):
    no_of_country = [y_axis.count(level) for level in x_axis]
    cumulative_freq = np.cumsum(no_of_country)
    
    plt.figure(figsize = (6, 4))
    plt.plot(x_axis, cumulative_freq, zorder = 3)
    plt.scatter(x_axis, cumulative_freq, marker='o', color = 'green', alpha = 0.9, zorder = 4)
    plt.fill_between(x_axis, cumulative_freq, alpha=0.3,color = 'pink', zorder = 2)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.grid(axis = 'both', alpha = 0.2, linestyle = ':', zorder = 1)
    for i, v in enumerate(cumulative_freq):
        plt.text(i, v, str(v), ha='right', va='bottom', fontsize = 8)
        
    #plt.savefig(title+'.svg')
    plt.show()

#purchasing power aspect
OgiveGraph("Ogive of Countries Vs. Purchasing Power Category", "Purchasing POwer Category", "Cumulative Number of Countries", levels, purchasing_power_category)

#safety aspect
OgiveGraph("Ogive of Countries Vs. Safety Category", "Safety Category", "Cumulative Number of Countries", levels, safety_category)

#health care aspect
OgiveGraph("Ogive of Countries Vs. Health Care Category", "Health Care Category", "Cumulative Number of Countries", levels, health_care_category)

#Cost of Living aspect
OgiveGraph("Ogive of Countries Vs. Cost of Living Category", "Cost of Living Category", "Cumulative Number of Countries", levels, col_category)

#pollution aspect
OgiveGraph("Ogive of Countries Vs. Pollution Category", "Pollution Category", "Cumulative Number of Countries", levels, pollution_category)
```

```python
#plot for pie chart
def PieChart(title, x_axis, y_axis):
    no_of_country = [y_axis.count(level) for level in x_axis]
    
    plt.figure(figsize=(8, 5))
    plt.pie(no_of_country, labels=x_axis, autopct='%1.1f%%', startangle=50,)
    plt.axis('equal')
    plt.title("Distribution of Countries by " + title)
    plt.legend(x_axis, title=title, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
    plt.tight_layout()
    #plt.savefig("Distribution of Countries by " + title+'.svg')
    plt.show()

#purchasing power aspect
PieChart("Purchasing Power Category",levels,purchasing_power_category)

#safety aspect
PieChart("Safety Category",levels,safety_category)

#health care aspect
PieChart("Health Care Category",levels,health_care_category)

#Cost of Living aspect
PieChart("Cost of Living Category",levels,col_category)

#pollution aspect
PieChart("Pollution Category",levels,pollution_category)
```

```python
#plot for Box plots
def BoxPlot(title, x_label, y_label, x_axis, value_list, category_list,):

    grouped_value = [
        [score for score, category in zip(value_list, category_list) if category == level]
        for level in x_axis
    ]
    
    plt.figure(figsize = (8, 8))
    plt.boxplot(grouped_value, labels=x_axis, patch_artist=True, )
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.grid(True,axis = 'both', linestyle = '--', alpha = 0.5)
    #plt.savefig(title+'.svg')
    #plt.yticks(np.arange(0, 101, 5))
    #plt.ylim(0, 100)
    plt.show()
   
#purchasing power aspect 
purchasing_value = data["Purchasing Power Value"].tolist()
BoxPlot("Box Plot of Purchasing Scores by Category", "Purchasing Power Category", 'Purchasing Scores', levels, purchasing_value, purchasing_power_category,)
   
#safety aspect
safety_value = data["Safety Value"].tolist()
BoxPlot("Box Plot of Safety Scores by Category", "Safety Category", 'Safety Scores', levels, safety_value, safety_category,)

#health care aspect
health_value = data["Health Care Value"].tolist()
BoxPlot("Box Plot of Health Scores by Category", "Health Care Category", 'Health Scores', levels, health_value, health_care_category,)

#Cost of Living aspect
col_value = data["Cost of Living Value"].tolist()
BoxPlot("Box Plot of Cost of Living Scores by Category", "Cost of Living Category", 'Cost of Living Scores', levels, col_value, col_category,)

#pollution aspect
pollution_value = data["Pollution Value"].tolist()
BoxPlot("Box Plot of Pollution Scores by Category", "Pollution Category", 'Pollution Scores', levels, pollution_value, pollution_category,)
```

```python
"""for purchasing power"""

mean_continents_purchasing_power = {}
median_continents_purchasing_power = {}
quartiles_Q1_continets_purchasing_power = {}
quartiles_Q3_continets_purchasing_power = {}
std_continents_purchasing_power = {}

for continent in continents:
    #all operations are being done using this list, hence is computed only once
    operating_list = list(data.loc[[index for index, country in zip(data.index, data["country"]) if country in continents[continent]]]["Purchasing Power Value"])
    mean_continents_purchasing_power[continent] = np.mean(operating_list)
    median_continents_purchasing_power[continent] = np.median(operating_list)
    quartiles_Q1_continets_purchasing_power[continent] = np.percentile(operating_list, 25)
    quartiles_Q3_continets_purchasing_power[continent] = np.percentile(operating_list, 75)
    range_purchasing_power = np.max(operating_list) - np.min(operating_list)
    std_continents_purchasing_power[continent] = stats.stdev(operating_list)
    mean_dev = np.mean(np.abs(operating_list - mean_continents_purchasing_power[continent]))
    
    print("\t\tFor", continent+ "'s Purchasing Power Values",)
    print(f"The average is {mean_continents_purchasing_power[continent]:.3f}")
    print(f"The median is {median_continents_purchasing_power[continent]:.3f}")
    print(f"Standard Dev is {std_continents_purchasing_power[continent]:.3f}")
    print(f"Mean dev is {mean_dev:.3f}")
    print(f"1st Quartile is {quartiles_Q1_continets_purchasing_power[continent]:.3f}")
    print(f"2nd Quartile is {median_continents_purchasing_power[continent]:.3f}")
    print(f"3rd Quartile is {quartiles_Q3_continets_purchasing_power[continent]:.3f}")
    print(f"range is {range_purchasing_power:.3f}\n")
```

```python
def CentralLimitTheorem(title,operating_list, sample_size_list, no_of_samples):
    #for each sample size plot the graph as below sideways such that each row contains 3 graphs
    nrows = int(np.ceil(len(sample_size_list) / 2))
    fig,ax = plt.subplots(nrows=nrows,ncols=2, figsize = (15, 5 * nrows))
    fig.suptitle(title, fontsize=16)
    ax = ax.flatten()
    for i,sample_size in enumerate(sample_size_list) :
        sample_means = []
        for _ in range(no_of_samples):
            sample = np.random.choice(operating_list, sample_size)
            sample_means.append(np.mean(sample))
        sample_means_dev = stats.stdev(sample_means)
        sample_means_mean = np.mean(sample_means)
        
        x = np.linspace(min(sample_means), max(sample_means), 100)
        y = (1 / (sample_means_dev * np.sqrt(2 * np.pi))) * np.exp(-((x - sample_means_mean) ** 2) / (2 * sample_means_dev ** 2))
        
        ax[i].hist(sample_means, bins=20, color = 'brown', edgecolor='black', alpha = 0.1, rwidth=1, density=True, label= "Approximation by Histograms")
        sns.kdeplot(sample_means,ax = ax[i], color='green', linewidth=2, label='Estimated Distribution of Sample Means')
        ax[i].plot(x, y, color='red', linestyle='dashed', label='Approximation by Normal Distribution')
        ax[i].set_title(f"Sample Size = {sample_size}")
        ax[i].set_xlabel("Sample Means")
        ax[i].set_ylabel("Frequency")
        ax[i].legend()
        
    print(f"{title} mean is {sample_means_mean:.3f} and standard deviation is {sample_means_dev:.3f} which follows a normal distribution\n")
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    #plt.savefig(title+".svg")
    plt.show()
        
        
sample_sizes = [1, 10, 50, 120,]
no_of_samples = 1000

CentralLimitTheorem("Purchasing Power Values",purchasing_value, sample_sizes, no_of_samples)
CentralLimitTheorem("Safety Values",safety_value, sample_sizes, no_of_samples)
CentralLimitTheorem("Health Care Values",health_value, sample_sizes, no_of_samples)
CentralLimitTheorem("Cost of Living Values",col_value, sample_sizes, no_of_samples)
CentralLimitTheorem("Pollution Values",pollution_value, sample_sizes, no_of_samples)
```
